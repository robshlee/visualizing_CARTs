# Visualizing CARTs with Adminssions Data
Using the provided admissions data, we'll build CARTs, look at how they work visually, and compare their performance to more standard, parametric models. As the focus for this exercise is comprehension, I'll ignore normal MOs like train test splitting.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats

from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier

from ipywidgets import *
from IPython.display import display

plt.style.use('fivethirtyeight')
%matplotlib inline

# graphviz and pydotplus are visualization libraries that will aid
# in understanding decision trees
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
```

```python
df = pd.read_csv('./admissions.csv')

print(df.shape)
df.head()
```

```python
df.isnull().sum()
```

```python
# Drop Nans
df.dropna(inplace=True)
```

We'll create 2 targets, admit and gpa. We'll do this so we can run regression models in addition to classification models for decision trees.

```python
# Set regression and classification X, y data
Xr = df[['admit', 'gre', 'prestige']]
yr = df['gpa']

Xc = df[['gpa', 'gre', 'prestige']]
yc = df['admit']
```

### Regression
We'll first run a linear model and see how it performs with the regression data.

```python
scores = cross_val_score(LinearRegression(), Xr, yr, cv=5)
print('Crossfold mean:', scores.mean())
print('Crossfold std:', scores.std())
```

Our linear model is performing pretty poorly. Let's try building a few decisin trees.

```python
def build_tree(reg_or_clas, max_depth):
    '''This function builds a decision tree regressor or classifier with 
    a provided max depth. It uses the 'X' and 'y' defined outside the function
    and prints the cross validation score with 5 folds.'''
    if reg_or_clas == 'reg':
        dtr = DecisionTreeRegressor(max_depth=max_depth)
        dtr.fit(Xr, yr)
        score = cross_val_score(dtr, Xr, yr, cv=5).mean()
        print('Max depth: {} with score {}'.format(max_depth, score))
    elif reg_or_clas == 'clas':
        dtc = DecisionTreeClassifier(max_depth=max_depth)
        dtc.fit(Xc, yc)
        score = cross_val_score(dtc, Xc, yc, cv=5).mean()
        print('Max depth: {} with score {}'.format(max_depth, score))
```

```python
build_tree('reg', 1)
build_tree('reg', 2)
build_tree('reg', 3)
build_tree('reg', None)
```

Overall the cross validated scores are all bad. Let's graphically see what our decision trees are doing.

```python
# Max depth = 1
dtr = DecisionTreeRegressor(max_depth=1)
dtr.fit(Xr, yr)

dot_data = StringIO()
export_graphviz(dtr, out_file=dot_data, filled=True, rounded=True, \
               special_characters=True, feature_names=Xr.columns)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

#### Interpreting a regression tree diagram

- First line is the condition used to split that node (go left if true, go right if false)
- `samples` is the number of observations in that node before splitting
- `mse` is the mean squared error calculated by comparing the actual response values in that node against the mean response value in that node
- `value` is the mean response value in that node

```python
# Max depth = 2
dtr = DecisionTreeRegressor(max_depth=2)
dtr.fit(Xr, yr)

dot_data = StringIO()
export_graphviz(dtr, out_file=dot_data, filled=True, rounded=True, \
               special_characters=True, feature_names=Xr.columns)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

Seems like there are too many decisions if we let it run on its own.

### Classification

We'll first build a logistic regression to compare our decision tree to.

```python
scores = cross_val_score(LogisticRegression(), Xc, yc, cv=5)
print('Crossfold mean:', scores.mean())
print('Crossfold std:', scores.std())
```

~71% cross validated score isn't too bad. Let's see how the decision trees compare.

```python
build_tree('clas', 1)
build_tree('clas', 2)
build_tree('clas', 3)
build_tree('clas', None)
```

With max depth of 3 seems to be marginally better than the logit model while all other depths performed slightly worse. Let's graphically depict the trees to peak at what's going on under the hood.
